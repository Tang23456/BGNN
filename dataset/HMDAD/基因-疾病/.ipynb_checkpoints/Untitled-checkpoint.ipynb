```python
import pandas as pd
import numpy as np

# 读取Excel文件
data = pd.read_excel('mapped_data.xlsx')

# 拼接MeSH ID和Disease列作为索引
data['Disease_Index'] = data['MeSH ID'] + '_' + data['Disease']

# 获取唯一的疾病和化学物质ID列表
diseases = data['Disease_Index'].unique()
chemicals = data['GeneIDs'].str.split(';').explode().unique()

# 创建疾病和化学物质的编码字典
disease_to_index = {disease: idx for idx, disease in enumerate(diseases)}
chemical_to_index = {chemical: idx for idx, chemical in enumerate(chemicals)}

# 初始化邻接矩阵A
A = np.zeros((len(diseases), len(chemicals)), dtype=int)

# 填充邻接矩阵A
for _, row in data.iterrows():
    disease_idx = disease_to_index[row['Disease_Index']]
    chemical_ids = row['GeneIDs'].split(';')
    for chemical_id in chemical_ids:
        chemical_idx = chemical_to_index[chemical_id]
        A[disease_idx, chemical_idx] = 1

# 计算度矩阵D
D = np.diag(A.sum(axis=1))

# 计算D的逆的平方根
D_inv_sqrt = np.linalg.inv(np.sqrt(D))

# 计算疾病相似度矩阵S
S = D_inv_sqrt @ A @ A.T @ D_inv_sqrt

# 将疾病相似度矩阵保存到CSV文件
pd.DataFrame(S, index=diseases, columns=diseases).to_csv('disease_similarity_matrix.csv')
```

```python
# 可视化（可选）
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(S, xticklabels=diseases, yticklabels=diseases, cmap='viridis')
plt.title('Disease Similarity Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取相似度矩阵文件
disease_similarity_matrix = pd.read_csv('./disease_similarity_matrix.csv', index_col=0)

# 获取疾病名称列表
diseases = disease_similarity_matrix.index

# 提取去重后的疾病名称列表
unique_disease_names = list(set(d.split('_')[1] for d in diseases))

# 创建一个空的 M x N 中间矩阵，数据类型为 float
intermediate_matrix = pd.DataFrame(0.0, index=diseases, columns=unique_disease_names)

# 填充 M x N 中间矩阵
for disease in diseases:
    disease_name = disease.split('_')[1]
    for unique_name in unique_disease_names:
        # 找到同名疾病的相似度
        same_named_diseases = [d for d in diseases if d.split('_')[1] == unique_name]
        if same_named_diseases:
            # 选取当前疾病与同名疾病中的最大相似度
            max_similarity = disease_similarity_matrix.loc[disease, same_named_diseases].max()
            intermediate_matrix.loc[disease, unique_name] = max_similarity

# 创建一个空的 N x N 最终矩阵，数据类型为 float
final_matrix = pd.DataFrame(0.0, index=unique_disease_names, columns=unique_disease_names)

# 填充 N x N 最终矩阵
for disease_name in unique_disease_names:
    for unique_name in unique_disease_names:
        # 选取当前疾病名称列与其他疾病名称列的最大相似度
        max_similarity = intermediate_matrix[unique_name].loc[[d for d in diseases if d.split('_')[1] == disease_name]].max()
        final_matrix.loc[disease_name, unique_name] = max_similarity

# 将处理后的相似度矩阵保存到CSV文件
final_matrix.to_csv('./processed_disease_similarity_matrix.csv')
```

```python
# 可视化矩阵
plt.figure(figsize=(10, 8))
sns.heatmap(final_matrix, cmap='viridis', annot=False, cbar=True)
plt.title('Disease Similarity Matrix')
plt.xlabel('Disease Names')
plt.ylabel('Disease Names')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
